```python
import pandas as pd
```

```python
ratings = pd.read_csv('ml-latest-small/ratings.csv')

ratings.head()
```

```python
movies = pd.read_csv('ml-latest-small/movies.csv')

movies.head()
```

```python
ratings.nunique()
```

```python
userID = ratings['userId'].unique()
movieID = movies['movieId']
```

```python
rating_data = {}
rating_list = []

for uID in userID:
    user = ratings[(ratings['userId'] == uID)]
    #print(uID)
    
    movie_index = 0
    for index, rating in user.iterrows():
        while rating['movieId'] != movieID[movie_index]: 
            movie_index += 1
            rating_list.append(0)
        
        rating_list.append(rating['rating'])
            
        movie_index += 1
        
    while movie_index < len(movieID):
        movie_index += 1
        rating_list.append(0)
    
    rating_data[uID] = rating_list
    rating_list = []
```

```python
df = pd.DataFrame(rating_data)

df = df.T
```

```python
df.columns = movieID
```

```python
df.to_csv('data/user_rating.csv', index=False)
```
