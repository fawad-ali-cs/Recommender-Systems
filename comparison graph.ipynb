```python
import matplotlib.pyplot as plt
```

```python
models = ['KNN User', 'KNN Item', 'Basic MF', 'MF with Bias']
RMSE = [1.010139, 0.889785, 0.895426, 0.854609]
```

```python
plt.bar(models, RMSE, color='darkred')

plt.title('Recommender System Comparison')
plt.xlabel('Recommender System', size=12)
plt.ylabel('Lowest RMSE')

plt.savefig('graphs/comparison.png')

plt.show()
```
